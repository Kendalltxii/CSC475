```python
import pandas as pd 

data = pd.read_csv('data/College_footabll_scores (2000-2020).csv')

data['Result'] = data['Result'].apply(lambda x: 1 if x == 'W' else 0)

furman = data[data['Team'].apply(lambda x: 'furman' in x.lower())]
wofford = data[data['Team'].apply(lambda x: 'wofford' in x.lower())]
furman['Date'] = pd.to_datetime(furman['Date'])
wofford['Date'] = pd.to_datetime(wofford['Date'])
furman['year'] = furman['Date'].dt.year
wofford['year'] = wofford['Date'].dt.year



# wofford['Result'] = pd.to_datetime(wofford['Date'])

avgs_furman = furman.groupby('year').mean()
avgs_wofford = wofford.groupby('year').mean()
```
