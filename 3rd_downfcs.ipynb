```python
import pandas as pd
from glob import glob 

fnames = glob ('data/cfbfcs21.csv')

all_data = []

for fname in fnames:
    data = pd.read_csv(fname) 
    data['year'] = fname.split('/')[1].split('.')[0][-2:]
    all_data.append(data)

all_data = pd.concat(all_data)

all_data.head()
```

```python
all_data.shape, data.shape
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(all_data['X3rd.Percent'], all_data['Win']/all_data['Games']);
plt.xlabel('3rd Down Conversion Percentage 2024');
plt.ylabel('Win Percentage');

plt.title('Win Percentage vs 3rd Down Conversion Percentage for \n FCS SoCon College Football Teams from 2023-2024');
```
